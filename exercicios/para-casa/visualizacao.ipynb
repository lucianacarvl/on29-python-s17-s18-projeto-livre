```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('base_final_tratada.csv')
df
```

```python
df['Data_Medicao'] = pd.to_datetime(df['Data_Medicao'])


df['Ano'] = df['Data_Medicao'].dt.year

df_agrupado = df.groupby('Ano')['PRECIPITACAO_TOTAL_MENSAL_AUT_mm_'].sum().reset_index()
```

```python
plt.figure(figsize=(10, 6))
plt.figure(figsize=(10, 6))
plt.plot(df_agrupado['Ano'], df_agrupado['PRECIPITACAO_TOTAL_MENSAL_AUT_mm_'], marker='o', linestyle='-', color='b')
plt.title('PrecipitaÃ§Ã£o Total Mensal por Ano')
plt.xlabel('Ano')
plt.ylabel('PrecipitaÃ§Ã£o Total (mm)')
plt.grid(True)
plt.xticks(df_agrupado['Ano'])
plt.tight_layout()
plt.show()
```

```python
df['Data_Medicao'] = pd.to_datetime(df['Data_Medicao'])
df['Ano'] = df['Data_Medicao'].dt.year
df_2024 = df[df['Ano'] == 2024].copy() 
df_2024.loc[:, 'MÃªs'] = df_2024['Data_Medicao'].dt.month
df_agrupado = df_2024.groupby('MÃªs')['PRECIPITACAO_TOTAL_MENSAL_AUT_mm_'].sum().reset_index()
```

```python
df['Data_Medicao'] = pd.to_datetime(df['Data_Medicao'])

df['Ano'] = df['Data_Medicao'].dt.year

df_filtrado = df[df['TEMPERATURA_MEDIA_MENSAL_AUT_C_'] > 0]

df_temperaturas = df_filtrado.groupby('Ano').agg({
    'TEMPERATURA_MEDIA_MENSAL_AUT_C_': ['max', 'min']
}).reset_index()

df_temperaturas.columns = ['Ano', 'Temperatura_Maxima', 'Temperatura_Minima']

print(df_temperaturas.head())
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_temperaturas['Ano'], df_temperaturas['Temperatura_Maxima'], marker='o', linestyle='-', color='red', label='Temperatura MÃ¡xima')
plt.plot(df_temperaturas['Ano'], df_temperaturas['Temperatura_Minima'], marker='o', linestyle='-', color='blue', label='Temperatura MÃ­nima')
plt.title('Temperaturas MÃ¡ximas e MÃ­nimas ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Temperatura MÃ©dia (Â°C)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
